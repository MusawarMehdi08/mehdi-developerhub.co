# encoding the data

```python
df.drop(['player no', 'player_name', 'team_abbreviation','college', 'country', 'season'],axis=1,inplace=True)
```

```python
df["draft_round"]=df["draft_round"].replace("Undrafted",int(0))
df["draft_number"]=df["draft_number"].replace("Undrafted",int(0))
df["draft_year"]=df["draft_year"].replace("Undrafted",int(0))
```

```python
df.info
```

```python
x=df.drop("pts",axis=1)
y=df["pts"]
print(x.shape)
```

```python
df
```

# train test split

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

# lineear regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(xtrain,ytrain)
```

```python
prediction=lr.predict(xtest)
```

```python
df=pd.DataFrame({"actual":ytest,"predict":prediction})
df.head(12)
```

```python
lr.score(xtest,ytest)
```

# ridge 

```python
from sklearn.linear_model import Ridge
rg=Ridge()
rg.fit(xtrain,ytrain)
rg.score(xtest,ytest)
```

```python
prediction1=rg.predict(xtest)
```

```python
df=pd.DataFrame({"actual":ytest,"predict":prediction1})
df.head(12)
```

```python
rg.score(xtest,ytest)
```

# evalulation of the model

```python
#on linear regression model
sme=mean_squared_error(ytest,prediction)
rsme=np.sqrt(sme)
print('mean_squared_error',sme)
print('root men sqaure error',rsme)
```

```python
# ridge model
sme=mean_squared_error(ytest,prediction1)
rsme=np.sqrt(sme)
print('mean_squared_error',sme)
print('root men sqaure error',rsme)
```

```python
#linear regression
r2square=r2_score(ytest,prediction)
print(r2square)
```

```python
#for ridge model
r2square=r2_score(ytest,prediction1)
print(r2square)
```
